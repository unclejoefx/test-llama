```python
import os
import pandas as pd
import dotenv
import litellm as llm

# Load environment variables from the .env file
dotenv.load_dotenv()

# Set the OpenAI API key from environment variables
if "OPENAI_API_KEY" in os.environ:
    OPENAI_API_KEY = os.environ['OPENAI_API_KEY']
else:
    raise EnvironmentError("OpenAI API key not found in environment variables.")

# Initialize litellm with your API key
llm.api_key = OPENAI_API_KEY

# Set the model name
Model = "gpt-4o-mini"

# Enable verbose mode for debugging
llm.set_verbose(True)

# Function to generate flight history using litellm's completion method
def generate_flight_history(prompt):
    response = llm.completion(
        model=Model,
        prompt=prompt,
        max_tokens=150,
        temperature=0.7
    )
    return response['choices'][0]['text'].strip()

# Sample prompt for generating flight history
prompt = """
Generate a realistic flight history for a frequent traveler named John Doe, who has taken several international and domestic flights in 2023. The details should include:
- Flight number
- Airline
- Departure date
- Departure time
- Arrival time
- Origin and destination cities

Here is the flight history:
"""

# Generate flight history using the LLM
try:
    generated_history = generate_flight_history(prompt)
    print("Generated Flight History:\n", generated_history)
except Exception as e:
    print(f"Error generating flight history: {e}")

# Mock data for demonstration (you need to parse generated text into a structured format)
flight_data = [
    {"Flight Number": "AA123", "Airline": "American Airlines", "Departure Date": "2023-01-15", "Departure Time": "14:30", "Arrival Time": "18:45", "Origin": "JFK", "Destination": "LAX"},
    {"Flight Number": "BA456", "Airline": "British Airways", "Departure Date": "2023-03-22", "Departure Time": "10:00", "Arrival Time": "12:30", "Origin": "LHR", "Destination": "CDG"},
]

# Convert the flight data to a pandas DataFrame
df_flight_history = pd.DataFrame(flight_data)

# Define the CSV file path
csv_file_path = "generated_flight_history.csv"

# Save the DataFrame to a CSV file
df_flight_history.to_csv(csv_file_path, index=False)

print(f"Flight history saved to {csv_file_path}")
```
